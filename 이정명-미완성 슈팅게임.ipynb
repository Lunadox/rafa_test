```python
'''
pygame과 기존의 이미지 파일을 써서 움직임 구현
좌우 방향키로 이동, 스페이스 키로 발사
'''

import pygame
### 1.게임 초기화
pygame.init()

### 2.게임창 옵션 설정
size = [400,900]
screen = pygame.display.set_mode(size) #변수화해서 저장

title ='P.G.'
pygame.display.set_caption(title)

### 3.게임내 필요한 설정
clock = pygame.time.Clock()

pygame.mixer.init()
pygame.mixer.music.load('If_I_Had_a_Chicken.mp3')
pygame.mixer.music.play(-1)
s_sound = pygame.mixer.Sound('boing.wav')

class obj:#주인공, 적군, 미사일을 표현할 물체 선언
    def __init__(self): #초기화
        self.x = 0
        self.y = 0
        self.move = 0 #움직임
    def put_img(self, address): #이미지를 가져오는 정의
        if address[-3:]=='png': # png파일만 컨버트
            self.img = pygame.image.load(address).convert_alpha()
        else:
            self.img = pygame.image.load(address)
        self.sx , self.sy =self.img.get_size() #받아온 이미지의 크키만큼을 선언
    
    def change_size(self,sx,sy):#받아온 이미지의 크기를 조절하는 함수
        self.img=pygame.transform.scale(self.img,(sx,sy))
        self.sx , self.sy =self.img.get_size()
        
    def show(self): #이미지 출력
        screen.blit(self.img,(self.x,self.y))
        
ss=obj()
ss.put_img('kitkat.gif')
ss.change_size(100,100)
ss.x=round(size[0]/2)-ss.sx/2
ss.y=size[1]-10-ss.sy
ss.move=5

left_go = False
right_go = False
space_go = False

m_list= []

black = (0,0,0)
white = (255,255,255)
k=0


### 4.메인 이벤트
SB=0
while SB==0:#0이 아니게 되면 종료
    
### 4-1. FPS 설정
    clock.tick(60)#FPS를 60
    
### 4-2. 각종 입력 감지
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            SB = 1
        
        if event.type==pygame.KEYDOWN: # 키를 계속 누르고 있다면
            if event.key == pygame.K_LEFT:
                left_go = True
            elif event.key == pygame.K_RIGHT:
                right_go = True
            elif event.key == pygame.K_SPACE:
                space_go = True
                k = 0
                
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_go = False
            elif event.key == pygame.K_RIGHT:
                right_go = False
            elif event.key == pygame.K_SPACE:
                space_go = False
            
### 4-3. 입력, 시간에 따른 변화
     
    if left_go == True:
        ss.x -= ss.move
        if ss.x<=0:
            ss.x = 0
    elif right_go == True:
        ss.x += ss.move
        if ss.x >= size[0]-ss.sx:
            ss.x = size[0]-ss.sx
    
   
    if space_go == True and k%6==0:
        
        s_sound.play()
        mm=obj() #미사일 선언
        mm.put_img('jellybean.gif')
        mm.change_size(80,80)
        mm.x=round(ss.x +ss.sx/2-mm.sx/2)
        mm.y=ss.y - mm.sy - 10
        mm.move=15
        m_list.append(mm)
    
    k += 1
    d_list = []
    for i in range(len(m_list)):

        m = m_list[i]
        m.y-=m.move
        if m.y <= -m.sy:
            d_list.append(i)
            
    for d in d_list:
        del m_list[d]
        
### 4-4. 그리기
    screen.fill(black)
    ss.show()
    for m in m_list:
        m.show()
### 4-5. 업데이트
    pygame.display.flip()

### 5. 게임종료
pygame.quit()
```
